# Product Performance

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_amazon_data.csv')
df.sample(3)
```

## Which product categories have the highest sales or ratings?

```python
# Split the 'category' column into separate subcategory columns
subcategory_cols = df['category'].str.split('|', expand=True)

# Rename the columns dynamically
subcategory_cols.columns = [f"subcategory_{i+1}" for i in range(subcategory_cols.shape[1])]

#Concatenate the new subcategory columns with the original DataFrame
df = pd.concat([df, subcategory_cols], axis=1)

df.sample(3)
```

```python
# Group by main category (subcategory_1) and calculate total rating count and average rating
main_category_performance = df.groupby('subcategory_1').agg({
    'rating_count': 'sum',
    'rating': 'mean'
}).reset_index()

top_sales_main_categories = main_category_performance.sort_values(by='rating_count', ascending=False)
top_rated_main_categories = main_category_performance.sort_values(by='rating', ascending=False)

print("Top Main Categories by Sales (Rating Count):")
print(top_sales_main_categories.head())

print("\nTop Main Categories by Ratings:")
print(top_rated_main_categories.head())
```

```python
# Group by subcategory_2 and calculate total rating count and average rating
subcategory_performance = df.groupby('subcategory_2').agg({
    'rating_count': 'sum',
    'rating': 'mean'
}).reset_index()

top_sales_subcategories = subcategory_performance.sort_values(by='rating_count', ascending=False)
top_rated_subcategories = subcategory_performance.sort_values(by='rating', ascending=False)

print("Top Subcategories by Sales (Rating Count):")
print(top_sales_subcategories.head())

print("\nTop Subcategories by Ratings:")
print(top_rated_subcategories.head())
```

## Visualize the Results

### Top Main Categories by Sales

```python
# # Top 10 main categories by sales
plt.figure(figsize=(12, 6))
sns.barplot(
    x='rating_count', 
    y='subcategory_1', 
    hue='subcategory_1',  # Assign y variable to hue
    data=top_sales_main_categories.head(10), 
    palette='viridis',
    legend=False  # Disable legend
)
plt.title('Top 10 Main Categories by Sales (Rating Count)')
plt.xlabel('Total Rating Count (Sales Proxy)')
plt.ylabel('Main Category')
plt.show()
```

### Top Main Categories by Ratings

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='rating', 
    y='subcategory_1', 
    hue='subcategory_1',
    data=top_rated_main_categories.head(10), 
    palette='magma',
    legend=False
)
plt.title('Top 10 Main Categories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Main Category')
plt.show()
```

### Top Subcategories by Sales

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='rating_count', 
    y='subcategory_2',
    hue = 'subcategory_2',
    data=top_sales_subcategories.head(10), 
    palette='viridis',
    legend=False
)
plt.title('Top 10 Subcategories by Sales (Rating Count)')
plt.xlabel('Total Rating Count (Sales Proxy)')
plt.ylabel('Subcategory')
plt.show()
```

### Top Subcategories by Ratings

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='rating', 
    y='subcategory_2', 
    hue='subcategory_2',
    data=top_rated_subcategories.head(10), 
    palette='magma',
    legend=False
)
plt.title('Top 10 Subcategories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Subcategory')
plt.show()
```
